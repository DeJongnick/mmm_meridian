```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("../data/raw/data.csv")

df['Date'] = pd.to_datetime(df['Date'])
df['Geo'] = 'Paris'

df.to_csv("../data/processed/data_processed.csv", index=False)
display(df.head())
```

```python
fig = px.line(df, x='Date', y='Sales', title='Sales over Time')
fig.show()
```
